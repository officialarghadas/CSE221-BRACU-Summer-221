```python
def dfs(n, l, v, s):
    v[n] = 1

    for adj_n in l[n]:
        if v[adj_n] == 0:
            dfs(adj_n, l, v, s)

    s.append(n)

def dfs_t(l_t, v_t, src):
    v_t[src] = 1
    print(src, end=" ")
    for adj_n in l_t[src]:
        if v_t[adj_n] == 0:
            dfs_t(l_t, v_t, adj_n)

with open("input3.txt", "r") as f:
    n, m = map(int, f.readline().split())
    l = [[] for i in range(n+1)]
    l_t = [[] for i in range(n+1)]
    v = [0] * (n + 1)
    v_t = [0] * (n + 1)
    s = []

    for i in range(m):
        u_v, v_v = map(int, f.readline().split())  
        l[u_v].append(v_v)
        l_t[v_v].append(u_v)

    for i in range(1, n + 1):
        if v[i] == 0:
            dfs(i, l, v, s)

    c_o = []
    while s:
        c_o.append(s.pop())

    print(c_o)
    for i in c_o:
        if v_t[i] == 0:
            dfs_t(l_t, v_t, i)
            print()

with open("output3.txt", "w") as f:
    for i in c_o:
        f.write(str(i) + " ")
    f.write("\n")
    for i in c_o:
        if v_t[i] == 0:
            dfs_t(l_t, v_t, i)
            f.write("\n")
```
