```python
def DFS(n, l, v, s):
    v[n] = 1

    for adj_n in l[n]:
        if v[adj_n] == 0:
            DFS(adj_n, l, v, s)

    s.append(n)

def TopoLogical_Sort(n, graph):
    in_degree = [0] * (n + 1)
    result = []

    for i in range(1, n + 1):
        for adj_node in graph[i]:
            in_degree[adj_node] += 1

    s = []
    for i in range(1, n + 1):
        if in_degree[i] == 0:
            s.append(i)

    while s:
        node = s.pop()
        result.append(node)

        for adj_node in graph[node]:
            in_degree[adj_node] -= 1
            if in_degree[adj_node] == 0:
                s.append(adj_node)

    return result

with open("input1a.txt", "r") as f:
    n, m = map(int, f.readline().split())

    graph = {}
    for i in range(1, n + 1):
        graph[i] = []

    for i in range(m):
        u, v = map(int, f.readline().split())
        graph[u].append(v)

    order = TopoLogical_Sort(n, graph)

if len(order) == n:
    with open("output1a.txt", "w") as f:
        f.write(" ".join(str(node) for node in order))
else:
    with open("output1a.txt", "w") as f:
        f.write("IMPOSSIBLE")
```
