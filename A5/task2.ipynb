```python
def dfs(n, l, visited, stack):
    visited[n] = 1

    for adj_n in l[n]:
        if visited[adj_n] == 0:
            dfs(adj_n, l, visited, stack)

    stack.append(n)

def h_cycle(n, l):
    visited = [0] * (n + 1)
    rec_stack = [0] * (n + 1)

    for i in range(1, n + 1):
        if visited[i] == 0 and i_cyclic(i, l, visited, rec_stack):
            return True

    return False

def i_cyclic(n, l, visited, rec_stack):
    visited[n] = 1
    rec_stack[n] = 1

    for adj_n in l[n]:
        if visited[adj_n] == 0:
            if i_cyclic(adj_n, l, visited, rec_stack):
                return True
        elif rec_stack[adj_n] == 1:
            return True

    rec_stack[n] = 0
    return False

with open("input2.txt", "r") as f:
    n, m = map(int, f.readline().split())

    l = [[] for i in range(n+1)]
    visited = [0] * (n + 1)
    stack = []

    for i in range(m):
        u_val, v_val = map(int, f.readline().split())
        l[u_val].append(v_val)

    # Sort the adjacency list lexicographically
    for i in range(1, n + 1):
        l[i].sort()

    if h_cycle(n, l):
        o_d = "IMPOSSIBLE"
    else:
        for i in range(1, n + 1):
            if visited[i] == 0:
                dfs(i, l, visited, stack)

        o_d = ""
        for i in range(len(stack) - 1, -1, -1):
            o_d += str(stack[i])
            if i > 0:
                o_d += " "

with open("output2.txt", "w") as f:
    f.write(o_d)
```
