```python
def bfs(node, l, v, st, qu):
    qu.append(node)
    v[node] = 1

    while qu:
        C_n = qu.pop(0)
        st.append(C_n)

        for adj_n in l[C_n]:
            if v[adj_n] == 0:
                qu.append(adj_n)
                v[adj_n] = 1

def topo_sort(n, graph):
    in_d = [0] * (n + 1)
    result = []

    for i in range(1, n + 1):
        for adj_n in graph[i]:
            in_d[adj_n] += 1

    queue = []
    for i in range(1, n + 1):
        if in_d[i] == 0:
            queue.append(i)

    while queue:
        node = queue.pop(0)
        result.append(node)

        for adj_n in graph[node]:
            in_d[adj_n] -= 1
            if in_d[adj_n] == 0:
                queue.append(adj_n)

    return result

with open("input1b.txt", "r") as f:
    n, m = map(int, f.readline().split())

    graph = {}
    for i in range(1, n + 1):
        graph[i] = []

    for i in range(m):
        u, v = map(int, f.readline().split())
        graph[u].append(v)

    order = topo_sort(n, graph)

if len(order) == n:
    with open("output1b.txt", "w") as f:
        for i in range(len(order)):
            f.write(str(order[i]))
            if i < len(order) - 1:
                f.write(" ")
else:
    with open("output1b.txt", "w") as f:
        f.write("IMPOSSIBLE")
```
